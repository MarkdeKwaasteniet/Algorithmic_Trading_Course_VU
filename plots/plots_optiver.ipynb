```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_delta = pd.read_csv("delta_20_02.csv",header=None, index_col=1, names=["delta_hedge"])
df_delta = df_delta.dropna(axis=0)
df_pnl = pd.read_csv("pnl_20_02.csv",header=None, index_col=1, names=["total_pnl"])
df_pnl = df_pnl.dropna(axis=0)
df_volume = pd.read_csv("volume_20_02.csv",header=None, names=["volume_to_hedge"])
df_volume = df_volume.dropna(axis=0)
```

```python
plt.style.use('dark_background')
top_plt = plt.subplot2grid((5,4), (0, 0), rowspan=3, colspan=4)
top_plt.plot(range(len(df_pnl)),df_pnl['total_pnl'])
plt.title('Time Series of the PnL (2 hours)')
bottom_plt = plt.subplot2grid((5,4), (3,0), rowspan=1, colspan=4)
bottom_plt.bar(range(len(df_volume)), df_volume["volume_to_hedge"])
plt.title('\n Trading Volume in BMW', y=-0.60)
plt.gcf().set_size_inches(16,8)
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(16, 8)
# plt.plot(df_pnl.index,df_pnl["total_pnl"])
plt.plot(range(len(df_pnl.iloc[:500])), df_pnl["total_pnl"].iloc[:500])
plt.title("Time Series of the PnL")
plt.ylabel("PnL")
plt.xlabel("Time")
plt.yticks(fontsize=8)
plt.xticks(fontsize=8)
fig.autofmt_xdate()
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(16, 8)
plt.bar(range(len(df_volume.iloc[:500])),df_volume["volume_to_hedge"].iloc[:500])
plt.title("Time Series of the Volume")
plt.ylabel("Volume")
plt.xlabel("Time")
plt.yticks(fontsize=8)
plt.xticks(fontsize=8)
fig.autofmt_xdate()
plt.show()
```
